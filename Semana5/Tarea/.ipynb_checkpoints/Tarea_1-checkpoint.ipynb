```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import pandas as pd

from pandasql import sqldf
from pandasql import *
```

Copiamos un ejemplo igual para ver como funciona

```python
from pandasql import load_meat, load_births
meat = load_meat()
births = load_births()
```

```python
#mostramos los datos 
meat.head(2)
```

```python
births.head(2)
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
q = """
SELECT
 m.date, m.beef, c.births
 
FROM
 meat m

LEFT JOIN
 births c
 ON m.date = c.date

WHERE
 m.date > '1974-12-31';

"""
```

```python
df = pysqldf(q)
df.births = df.births.fillna(method='backfill')
```

```python
df.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df['beef'].rolling(12).mean(), color='b')
ax1.set_xlabel('months since 1975')
ax1.set_ylabel('cattle slaughtered', color='b')

ax2 = ax1.twinx()
ax2.plot(df['births'].rolling(12).mean(), color='r')
ax2.set_ylabel('babies born', color='r')
plt.title("Beef Consumption and the Birth Rate")
plt.show();
```
