```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
def f_1(x):
    a = []
    for i in range(len(x)):
        if x[i]%2 == 0:
            a.append(x[i]**2)
        else:
            a.append(mt.log(x[i]))
    return np.array(a)
```

```python
def g(x):
    if x > 0:
        return np.cos(x)
    if x < 0:
        return np.sin(x)
    
g_1 = np.vectorize(g)
```

```python
x_1 = np.linspace(-2*np.pi, 2*np.pi,50)
y_1 = g_1(x_1)
```

```python
plt.plot(x_1,y_1)
```

Forma 2 (importante)

```python
y_2 = np.zeros_like(x_1)
w = x_1 > 0
%timeit y_2[w] = g_1(x_1[w])
%timeit y_2[~w] = g_1(x_1[~w])
```

```python
plt.plot(x_1,y_2)
```

Forma 3

```python
y_3 =np.asarray([g(x) for x in x_1])
print (y_3)
%timeit y_3
```

```python
class funciones():
    '''Se definen mis funciones basicas 
    A: Corresponde a la amplitud para las funciones coseno'''
    
    # inicializar variables
    def __init__(self,A, saludo = 'Hola' ):  #self para indicar que es parte de la clase
        self.A = A
        self.saludo = saludo
    
    #definimos funciones
    def g(self,x):  # siempre poner self para indicar que es parte de la clase
        self.cosx = self.A*np.cos(x)
        self.sinx = self.A*np.sin(x)
        return self.cosx, self.sinx
    
    def f_1(self):    #self indica que solamenete se debe inicializar la clase
        return 'Hola'
    
    def f_2(self,x):
        if x%2 == 0:
            return x**2
        else:
            return mt.log(x)
        
    def f_3(self):
        return self.saludo
```

```python
x = funciones(3) # 3 es el valor de A
```

```python
x.f_1()
```

```python
x.f_3()
```

```python
x.g(np.pi)
```

```python
x.f_2(4)
```

Descargar datos de una página HTML

```python
import requests
import pandas as pd
import os
```

```python
file = 'salarios_minimos.csv'
if not os.path.isfile(file):
    url = 'https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007'
    html = requests.get(url).content
    df_list = pd.read_html(html,header=1)
    df = df_list[-1]
    df.to_csv('salarios_minimos.csv')
else:
    df = pd.read_csv(file)

df
```
