## Flujo de Banda Dereddened en X-Y

```python
dered_flux = [data.dered_flux_g,data.dered_flux_r,data.dered_flux_z]
na_dered_flux = ['Dereddened G-band flux', 'Dereddened R-band flux' , 'Dereddened Z-band flux']

plt.title('Dereddened',fontsize=18)
for ax, fluxs, name in zip(axs.flat, dered_flux,na_dered_flux):
    cond_1 = fluxs < 1.0 
    cond_2 = fluxs > 0.0
    plt.scatter(data.bx[(cond_1) & (cond_2)], data.by[(cond_1) & (cond_2)], label=None,
            c = fluxs[(cond_1) & (cond_2)], s = 15, cmap='inferno', linewidth=0, alpha=1)
    plt.xlabel('x',fontsize=18)
    plt.ylabel('y',fontsize=18)
    plt.colorbar(label= name);
    plt.tight_layout()
    plt.show()
```
