```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits 
```

```python
path = '/home/echeveste/Mis_trabajos/analisis_datos/da2019-Oscar2401/ASTRO/survey-bricks-dr8-north.fits'
```

```python
hdul = fits.open(path)
```

```python
data1 = hdul[1].data
```

```python
plt.plot(data1['ra'],data1['dec'],',')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data1['ra'][::-1], data1['dec'], label=None,
            c=data1['nexp_g'], cmap='viridis', linewidth=0, alpha=1)
plt.axis(aspect='equal')
#plt.xlim(90,305)
#plt.ylim(30,80)
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label='Latitud');
```

```python
n = 0
ra = np.delete(data1['ra'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>n+5)))
```

```python
dec = np.delete(data1['dec'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>n+5)))
```

```python
nexp_g = np.delete(data1['nexp_g'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>n+5)))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ra[::-1], dec, label=None,
            c= nexp_g,cmap=plt.cm.get_cmap('nipy_spectral', 5), linewidth=0, alpha=1)
plt.axis(aspect='equal')
plt.xlim(85,310)
plt.ylim(30,80)
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label='Latitud');
```
