```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits 
```

## MzLS + BASS DR4: Number of exposures in g

```python
path = '/home/echeveste/Mis_trabajos/analisis_datos/da2019-Oscar2401/ASTRO/new_survey_bricks_dr4_2.fits'
hdul = fits.open(path)
data1 = hdul[1].data
```

```python
plt.plot(data1['ra'][::-1],data1['dec'],',')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data1['ra'][::-1], data1['dec'], label=None,
            c=data1['nexp_g'], cmap='viridis', linewidth=0, alpha=1)
plt.axis(aspect='equal')
plt.title('MzLS + BASS DR4: Number of exposures in g')
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label='Latitud');
```

#### We take only the number of exposures among 1 to 5

```python
n = 1
ra = np.delete(data1['ra'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>=n+6)))
dec = np.delete(data1['dec'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>=n+6)))
nexp_g = np.delete(data1['nexp_g'],np.where((data1['nexp_g']<n)|(data1['nexp_g']>=n+6)))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(ra[::-1], dec,
            c= nexp_g,cmap=plt.cm.get_cmap('viridis', 5), linewidth=0, alpha=1)
plt.axis(aspect='equal')
plt.title('MzLS + BASS DR4: Number of exposures in g')
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(ticks = range(n+5), label = 'Number of exposures in g')
plt.clim(n-0.5, n+4.5);
```
